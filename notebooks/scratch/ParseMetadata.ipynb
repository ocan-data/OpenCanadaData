# Parse StatscanMetadata

```python
import sys
import pandas as pd
from IPython.display import display
sys.path.append('..')
```

```python
def parse_sections(metadata_file):
    with open(metadata_file, 'r', encoding='utf8') as f:
        start_section = True
        for line in f.readlines():
            line = line.strip()
            if len(line) > 0:
                parts = [p.replace('"','') for p in line.split(',')]
                if start_section:
                    columns = parts
                    rows = []
                    start_section = False
                else:
                    rows.append(tuple(parts))
            else:
                start_section = True
                row_widths = [len(row) for row in rows]
                if row_widths == []:
                    continue
                row_width = max(row_widths)
                if row_width > len(columns):
                    rows = [row[:len(columns)] for row in rows]
                elif len(columns) > row_width:
                    columns = columns[:row_width]
                df = pd.DataFrame(data=rows, columns=columns)
                yield df

                
for df in parse_sections('../../data/23100274_MetaData.csv'):
    display(df)
```
