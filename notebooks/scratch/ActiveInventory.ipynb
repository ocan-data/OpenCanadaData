# Determine Active Inventory

```python
import sys

sys.path.append('../..')
```

# Load Inventory

```python
from ocandata import Inventory
inventory = Inventory()
```

# Get Inactive Urls

```python
inactive_urls = inventory.get_inactive_dataset_urls(inventory.data)
```

# Get Inactive Urls for Both Languages

```python
data = inventory._view().data
data = data.dropna(subset=['portal_url'])
data.portal_url = data.portal_url.astype(str)
active_data = data[~data.portal_url.isin(inactive_urls)]
inactive_data = data[data.portal_url.isin(inactive_urls)]
inactive_indexes = inactive_data.index.tolist()
inactive_indexes

french_data = inventory.FR.data
english_data = inventory.EN.data

inactive_urls_fr = french_data.iloc[inactive_indexes].portal_url.tolist()
inactive_urls_en = english_data.iloc[inactive_indexes].portal_url.tolist()
all_inactive_urls = list(set(inactive_urls_en + inactive_urls_fr))
all_inactive_urls = [url for url in all_inactive_urls if isinstance(url, str)]
```

# Save Inactive Urls

```python
with open('ExpiredDatasets.txt', 'w') as f:
    for url in all_inactive_urls:
        f.write(url)
        f.write('\n')
```
