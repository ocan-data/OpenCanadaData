```python
import pandas as pd
import numpy as np

pd.set_option('display.max_columns', 50)
```

```python
tables = pd.read_html('https://www.leaguelineup.com/teams_baseball.asp?url=eosba&teamid=5055885')
```

```python
batting = tables[0].rename(columns={'#': "Jersey"})
batting.Jersey = batting.Jersey.fillna(99).astype(int)
batting = batting[~(batting.Jersey==19)]
batting = batting[~(batting.Name == 'Total')]
batting=batting.drop(columns=['Jersey']).set_index('Name')
```

```python
def rnd(num):
    return np.round(num, 3)
batting['OB'] = batting.OBP * batting.PA
batting.OB = batting.OB.apply(np.round).astype(int)
batting['Walks'] = batting.OB - batting.H
batting['WalkPct'] = batting.Walks / batting.PA
batting.WalkPct = batting.WalkPct.apply(rnd)
batting['KPct'] = batting.K / batting.PA
batting.KPct = batting.KPct.apply(rnd)
batting['SLG'] = batting.TB / batting.PA
batting.SLG = batting.SLG.apply(rnd)
```

```python
batting.sort_values(['SLG'], ascending=False)
```

```python
batting.to_csv('../../batting.csv')
```

```python
class Filter:
    
    def __init__(self, df: pd.DataFrame, col: str):
        self.col = col
        self.df = df
        
    def __call__(self, name:str):
        return self.df[self.df.Name==name]
    
batting_stats = Filter(batting, 'Name')
batting_stats
```
